```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import date
from datetime import timedelta
```

The below code is adapted from timeline tutorial https://dadoverflow.com/2021/08/17/making-timelines-with-python/

```python
dates = [date(2001,8,30), date(2001,9,15), date(2002,2,2), date(2002,5,5), date(2002,9,15), date(2002,11,3), date(2003,12,31), date(2004,1,20), date(2004,2,17), date(2004,3,2), date(2004,4,16), date(2004,6,19)]
min_date = date(np.min(dates).year, np.min(dates).month - 3, np.min(dates).day)
max_date = date(np.max(dates).year , np.max(dates).month + 2, np.max(dates).day)
 
labels = ['Von Ryker Investigated\nfor Ethical Misconduct', 'Von Ryker Institute AG Closed', 'Luthor Announces Biotech\n Lab "Boynton Laboratories"\n to Locate in Alderwood.', 
          'Boynton Labs Groundbreaking', 'Boynton Laboratories Opens', 'Experts question reliability\n of meat import inspections', 'National rapid testing for BSE\n to occur at Boynton Labs', 'Boynton Lab scientists produce protein that can\n trigger disease in mice similar to "Mad Cow"', 'New Form of Mad Cow\n Disease Found in Italy', 'Boynton Labs announces development of treatment\n for diseases such as Mad Cow', 'FDA begins investigation of Boynton Labs' , 'FDA discontinues investigation\n of Boynton Labs']
# labels with associated dates
labels = ['{0:%d %b %Y}:\n{1}'.format(d, l) for l, d in zip (labels, dates)]
```

```python
fig, ax = plt.subplots(figsize=(6, 10), constrained_layout=True)
_ = ax.set_xlim(-20, 20)
_ = ax.set_ylim(min_date, max_date)
_ = ax.axvline(0, ymin=0.05, ymax=0.95, c='deeppink', zorder=1)
 
_ = ax.scatter(np.zeros(len(dates)), dates, s=120, c='palevioletred', zorder=2)
_ = ax.scatter(np.zeros(len(dates)), dates, s=30, c='darkmagenta', zorder=3)

label_offsets = np.repeat(2.0, len(dates))
label_offsets[1::2] = -2.0
for i, (l, d) in enumerate(zip(labels, dates)):
    d = d - timedelta(days=20)
    align = 'right'
    if i % 2 == 0:
        align = 'left'
    _ = ax.text(label_offsets[i], d, l, ha=align, fontfamily='serif', fontweight='bold', color='royalblue',fontsize=8)
stems = np.repeat(2.0, len(dates))
stems[1::2] *= -1.0   
x = ax.hlines(dates, 0, stems, color='darkmagenta')
# hide lines around chart
for spine in ["left", "top", "right", "bottom"]:
    _ = ax.spines[spine].set_visible(False)
 
# hide tick labels
_ = ax.set_xticks([])
_ = ax.set_yticks([])
 
_ = ax.set_title('Timeline of Events Relating to Boynton Labs', fontweight="bold", fontfamily='serif', fontsize=16, 
                 color='royalblue')
```
